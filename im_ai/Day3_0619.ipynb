# 기술 통계량 확인 (DDA)
- 데이터의 통계적 특성을 확인
- 연속형 : 
    - 대표값 (평균 mean / 중앙값 median) : 해당 데이터를 대표하는 값
        - 평균의 경우 이상치(outliar)에 의해 좌우될 수 있음
    - 산포 (분산 / 표준편차)
        - 대표값에 대해 데이터가 얼마나 정확한지 나타내는 척도
        - 데이터의 신뢰도를 나타내기도 함
    - 분포의 모양
        - 분포의 모양에 따라 가설검정 기법 / 회귀분석 방법이 달라질 수 있음
        - 가설검정을 통해 정규성 확인

- 범주형 : 항목 / 빈도수

```python
# 연령대 
# 연속형 데이터를 범주형 데이터로 만들어서 새로운 시각으로 보겠다는 목적
cond1 = (df1['Age']>=70)
cond2 = (df1['Age']>=60)
cond3 = (df1['Age']>=50)
cond4 = (df1['Age']>=40)
cond5 = (df1['Age']<40)

# .loc[조건, 생성할 칼럼] = '부여할 값'
df1.loc[cond5, '연령대'] = '40대 이하'
df1.loc[cond4, '연령대'] = '40대'
df1.loc[cond3, '연령대'] = '50대'
df1.loc[cond2, '연령대'] = '60대'
df1.loc[cond1, '연령대'] = '70대 이상'
```

```python
df1[['Age','연령대']]
```

```python
df1['연령대'].value_counts()
```

```python
# 함수 만들어서 노가다 말고 연령대로 나누기
def age_func(x):
    if x>= 70:
        return '70대 이상'
    elif x>= 60:
        return '60대'
    elif x>= 50:
        return '50대'
    elif x>= 40:
        return '40대 이하'

df1['Age'].apply(age_func) # age 항목의 모든 값을 for문 돌리듯이 함수 돌려준다
```

```python
df1['연령대2'] = df1['Age'].apply(age_func)
df1.head()
```

```python
# 또 다른 연령대 컬럼 만드는 방법
# age를 스트링으로 받아서 맨 앞글자 따와서 + *"0대" 붙임
d1 = df1['Age'][0]
type(d1)
```
