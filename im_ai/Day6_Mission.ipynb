## 2. '지점' 중 '판매량'의 합이 가장 높은 2개 지점을 확인하시오.

```python
d = df.pivot_table(index = '지점',
               values = '판매량',
               aggfunc = 'sum').reset_index()
d
top2 = d.sort_values(by = '판매량', ascending = False)['지점'].values[:2]
top2
```

## 3. 앞서 확인한, 2개의 지점의 '판매량'에 대해 각각 정규성 검정을 수행하고, 결론을 작성하시오. 

```python
cond1 = (df['지점'] == top2[0])
cond2 = (df['지점'] == top2[1])

dff = df.loc[cond1]
dfn = df.loc[cond2]

print(f"{top2[0]} : { stats.normaltest(dff['판매량']) }")
print(f"{top2[1]} : { stats.normaltest(dfn['판매량']) }")
# 결론 
# 부천점에서는 정규성을 따름 -> 귀무가설을 채택
# 안양점에서는 정규성을 따르지 x -> 대립가설을 채택
# 둘 중하나라도 정규성 따르지 않으면 ranksum함수 사용
```

## 4. 앞서 확인한, 2개의 지점의 '판매량'의 대표값(중앙값)의 차이를 계산하고자 한다. 가설을 수립하고, Rank Sum Test를 이용해 가설검정을 수행하시오 .

```python
# 순위합 검정 
stats.ranksums(dff['판매량'], dfn['판매량'])

# pvalue=0.8149463770358858
# p벨류 < 5 : 대립가설 채택
# 부천/안양의 판매량은 통계적으로 유의미한 차이가 있다
```

## 7. '상품명'에 '마스크'와'휴지' 제품을 나누어 분석하고자 한다. 상품명을 '휴지'와 '마스크'로 분류한 뒤, (나머지는 '기타') '상품분류'라는 파생변수를 만들어, 두 상품의 '판매량'의 대표값의 차이를 가설 검정 하시오. 
    - 가설 수립 및 결론도 작성하셔야 합니다.

```python
def func2(x):
    if '휴지' in x:
        return '휴지'
    elif '마스크' in x:
        return '마스크'
    else:
        return '기타'

df['상품분류'] = df['상품명'].apply(func2)
df

d = df.pivot_table(index = '상품분류', values = '판매량', aggfunc = 'sum').reset_index()

cond1 = (df['상품분류'] == '휴지')
cond2 = (df['상품분류'] == '마스크')

dff = df.loc[cond1]
dfn = df.loc[cond2]

print(f"휴지 : { stats.normaltest(dff['판매량']) }")
print(f"마스크 : { stats.normaltest(dfn['판매량']) }")
# 결론 
# 둘 중하나라도 정규성 따르지 않으면 ranksum함수 사용
```
